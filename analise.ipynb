```python
import pandas as pd
```

```python
dataset = pd.read_csv('restaurant.csv')
dataset
```

```python
dataset.dtypes
```

```python
price_interval = dataset['Price'].agg([min, max])
print("\nIntervalo (mínimo e máximo) da coluna 'Price':")
print(price_interval)

conjuntos_valores = dataset.apply(lambda col: set(col))
print("\nConjuntos de valores únicos de cada coluna:")
print(conjuntos_valores)
```

```python
num_instancias, num_atributos = dataset.shape

print(f"O dataset possui {num_instancias} instâncias")
```

```python
df = pd.DataFrame(dataset)

category_freq = df['Price'].value_counts()
price_range_freq = df['Profitability'].value_counts()

category_freq, price_range_freq
```

```python
hot_dataset = pd.get_dummies(dataset)
hot_dataset
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['Profitability'] = label_encoder.fit_transform(df['Profitability'])

df['Profitability']
```

```python
from sklearn.model_selection import train_test_split 

X = hot_dataset[hot_dataset.columns[:28]]
y = df['Profitability']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
import time
from sklearn import metrics
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

def train_and_evaluate_model(model):
    print(y_train)
    start_time = time.time()
    clf = model.fit(X_train, y_train)
    training_time = time.time() - start_time

    y_pred = model.predict(X_test)

    print("Tempo de treinamento:", training_time, "s\n")
    print("Acurácia:", metrics.accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred, target_names=['Low','Medium','High']))
    ConfusionMatrixDisplay.from_estimator(clf, X_test, y_test, display_labels=['Low','Medium', 'High'], xticks_rotation="vertical")
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.svm import SVC

train_and_evaluate_model(SVC(kernel='linear'))
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = train_and_evaluate_model(DecisionTreeClassifier(criterion='entropy',min_samples_leaf=1))
```

```python
from sklearn.ensemble import RandomForestClassifier

train_and_evaluate_model(RandomForestClassifier())
```
